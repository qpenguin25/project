```python
from peewee import *
from datetime import datetime

mysql_db = MySQLDatabase('ab104', host='localhost', user='root',passwd='Aa0931232013')

class BaseModel(Model):
    """A base model that will use our MySQL database"""
    class Meta:
        database = mysql_db

class News(BaseModel):
    id = IntegerField()
    title = CharField()
    url = CharField()
    content = TextField()
    pub_date = DateTimeField()
```
