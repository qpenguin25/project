```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
import time
import requests
from bs4 import BeautifulSoup
from Queue import Queue
from threading import Thread



Taiwan_area=[u'臺北',u'桃園',u'台中',u'台南',u'高雄',u'基隆',u'新竹',u'嘉義',u'苗栗',u'彰化',u'南投',u'雲林',u'屏東',u'宜蘭',u'花蓮',u'臺東',u'澎湖縣']
area_queue = Queue()
for area in Taiwan_area:
    area_queue.put(area)
    
    
####   爬蟲實施抓取網頁
def worker():
    while not area_queue.empty():
        crawler() 
        
        
####   爬蟲程序            
def crawler():    
    URL='https://tw.hotels.com/?intlid=SEARCHRESULTS+%3A%3A+header_main_section'
    driver = webdriver.Remote("http://localhost:9515", webdriver.DesiredCapabilities.CHROME)
    driver.get(URL)
    driver.find_element_by_xpath('//*[@id="qf-0q-destination"]').send_keys(area_queue.get())
    driver.find_element_by_xpath('//*[@id="main-content"]/main/div/div/div[1]/div/div[1]/div[1]/div/div/form/fieldset[5]/button').click()

    while True:
        try:
            driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
            time.sleep(1)
            driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
            soup = BeautifulSoup(driver.page_source,'lxml')
            item_list=soup.select('.hotel-wrap')
            print len(item_list)
            start=driver.find_element_by_css_selector('.unavailable-info')
            if start:
                print "找到了~~~~~~~~~~~"
                break
        except:
            print "沒找到. . ."

    item_list=soup.select('.hotel-wrap')
    for item in item_list:
        try:
            print item.select('.p-name')[0].text
            print item.select('.contact')[0].text
            print 'https://tw.hotels.com'+item.select('a')[0]['href']
            print item.select('.price > a > span > ins')[0].text
            print item.select('.price > a > span > del')[0].text
            print item.select('.location-info > ul ')[0].text
            print "---------------------------------------------------------------------"
        except Exception as e:
            print 'null'
            print e
            print "---------------------------------------------------------------------"
    if area_queue.empty():
        driver.close()
    else:
        driver.close()
        worker()
            
            
s1 = datetime.now() ####### 起始時間
####   啟動THREAD
threads = map(lambda i: Thread(target=worker), xrange(4))
map(lambda th: th.start(), threads)
map(lambda th: th.join(), threads)

s2 = datetime.now() ####### 結束時間
print "All  Finish "+str(s2-s1)+"!!" ####### 總共爬取資料耗費時間

```
